```python
def connect_to_existing_browser():
    chrome_options = Options()
    chrome_options.add_experimental_option("debuggerAddress", "127.0.0.1:9222")
    
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)
    return driver
```

```python
# 目标URL列表
input_file = 'C:\\Users\\Pratt\\Desktop\\HKUST-RA\\[-Ongoing-] Text analysis in Glassdoor data\\Page-Source-Scraping\\Page_Url_Formatting_Trans\\Urls-1.1.json'

try:
    # 读取JSON文件
    with open(input_file, 'r', encoding='utf-8') as f:
        company_data = json.load(f)
except FileNotFoundError:
    print(f"错误: 找不到文件 {input_file}")
    # 如果文件不存在，可以退出程序或使用默认数据
    raise
```

```python
for i, item in enumerate(company_data):
    # 跳过空URL
    if 'url' not in item or not item['url']:
        continue
    url = item['url']
    
    try:
        # 连接到已存在的浏览器
        driver = connect_to_existing_browser()
        
        # 登录信息
        # username = 'prattzan0618@gmail.com'
        # password = 'zanwenbo08062018'  

        driver.get(url)
        time.sleep(3)  # 等待页面加载
        
        # 保存数据
        pagesource = driver.page_source
        soup = BeautifulSoup(pagesource, 'html.parser')
        details_list = soup.find('ul', attrs={'data-test': 'companyDetails'})
        if details_list:
            for j, li in enumerate(details_list.find_all('li'), 1):
                text = li.get_text(strip=True)
                if text:
                    item[f'info-{j}'] = text
    except Exception as e:
        print(f"✗ 爬取失败: {e}")
        continue
```

```python
# 保存到新文件
output_file = 'C:\\Users\\Pratt\\Desktop\\HKUST-RA\\[-Ongoing-] Text analysis in Glassdoor data\\Page-Source-Scraping\\Page_Scrape_Check_Trans\\Urls-1.1.json'
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(company_data, f, ensure_ascii=False, indent=2)
```
